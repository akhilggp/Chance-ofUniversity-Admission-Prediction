# ANN

```python
import tensorflow
import keras
```

```python
admis= pd.read_csv('Admission_Predict.csv')
```

```python
admis
```

```python
X=admis.iloc[:,1:8].values
```

```python
X
```

```python
y=admis.iloc[:,8].values
```

```python
#Split the data into Training set and test set with 60% to training dataset and 40% to test dataseset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.4)
```

```python
#Perform feature scaling to fit and transform the training data using the StandardScaler function
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras 
from keras.models import Sequential     #Sequential module is used for initializing the neural network
from keras.layers import Dense          #Dense module to add the hidden layers.
```

```python
classifier = Sequential()

#Add the hidden layers one by one using the dense function.
classifier.add(Dense(units = 7, kernel_initializer = 'normal', 
activation = 'relu', input_dim = 7))

#Add second layer
classifier.add(Dense(units = 20, kernel_initializer = 'normal', 
activation = 'relu'))

#Add second layer
classifier.add(Dense(units = 1, kernel_initializer = 'normal', 
activation = 'linear'))

#Compile the ANN
classifier.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=['accuracy'])
#binary_crossentropy is same as cross validation..

classifier.fit(X_train, Y_train, batch_size = 10, epochs = 500)

y_pred_test= classifier.predict(X_test) 


y_pred_train = classifier.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rmse_training = (np.sqrt(mean_squared_error(Y_train, y_pred_train)))
r2_training = r2_score(Y_train, y_pred_train)

rmse_testing = (np.sqrt(mean_squared_error(Y_test, y_pred_test)))
r2_testing = r2_score(Y_test, y_pred_test)

# model evaluation for training set

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_training))
print('R2 score is {}'.format(r2_training))
print("\n")

# model evaluation for testing set

print("The model performance for testing set")
print("--------------------------------------")
print('Root mean squared error: ', rmse_testing)
print('R2 score: ', r2_testing)
```
